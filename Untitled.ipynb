```python
def get_channel_stats(youtube, channel_ids):
    """
    Get channel statistics: title, subscriber count, view count, video count, upload playlist
    Params:
    
    youtube: the build object from googleapiclient.discovery
    channels_ids: list of channel IDs
    
    Returns:
    Dataframe containing the channel statistics for all channels in the provided list: title, subscriber count, view count, video count, upload playlist
    
    """
    all_data = []
    request = youtube.channels().list(
                part='snippet,contentDetails,statistics',
                id=','.join(channel_ids))
    response = request.execute() 
    
    for i in range(len(response['items'])):
        data = dict(channelName   = response['items'][i]['snippet']['title'],
                    publishedDate = responsep['items'][i]['snippet']['publishedAt'],
                    subscribers   = response['items'][i]['statistics']['subscriberCount'],
                    views         = response['items'][i]['statistics']['viewCount'],
                    totalVideos   = response['items'][i]['statistics']['videoCount'],
                    playlistId    = response['items'][i]['contentDetails']['relatedPlaylists']['uploads'])
        all_data.append(data)
    
    return pd.DataFrame(all_data)

def get_video_ids(youtube, playlist_id):
    """
    Get list of video IDs of all videos in the given playlist
    Params:
    
    youtube: the build object from googleapiclient.discovery
    playlist_id: playlist ID of the channel
    
    Returns:
    List of video IDs of all videos in the playlist
    
    """
    
    request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId = playlist_id,
                maxResults = 50)
    response = request.execute()
    
    video_ids = []
    
    for i in range(len(response['items'])):
        video_ids.append(response['items'][i]['contentDetails']['videoId'])
        
    next_page_token = response.get('nextPageToken')
    more_pages = True
    
    while more_pages:
        if next_page_token is None:
            more_pages = False
        else:
            request = youtube.playlistItems().list(
                        part='contentDetails',
                        playlistId = playlist_id,
                        maxResults = 50,
                        pageToken = next_page_token)
            response = request.execute()
    
            for i in range(len(response['items'])):
                video_ids.append(response['items'][i]['contentDetails']['videoId'])
            
            next_page_token = response.get('nextPageToken')
        
    return video_ids

def get_video_details(youtube, video_ids):
    """
    Get video statistics of all videos with given IDs
    Params:
    
    youtube: the build object from googleapiclient.discovery
    video_ids: list of video IDs
    
    Returns:
    Dataframe with statistics of videos, i.e.:
        'channelTitle', 'title', 'description', 'tags', 'publishedAt'
        'viewCount', 'likeCount', 'favoriteCount', 'commentCount'
        'duration', 'definition', 'caption'
    """
        
    all_video_info = []
    
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_ids[i:i+50])
        )
        response = request.execute() 

        for video in response['items']:
            stats_to_keep = {'snippet': ['channelTitle', 'title', 'description', 'tags', 'publishedAt'],
                             'statistics': ['viewCount', 'likeCount', 'favouriteCount', 'commentCount'],
                             'contentDetails': ['duration', 'definition', 'caption']
                            }
            video_info = {}
            video_info['video_id'] = video['id']

            for k in stats_to_keep.keys():
                for v in stats_to_keep[k]:
                    try:
                        video_info[v] = video[k][v]
                    except:
                        video_info[v] = None

            all_video_info.append(video_info)


def get_comments_in_videos(youtube, video_ids):
    """
    Get top level comments as text from all videos with given IDs (only the first 10 comments due to quote limit of Youtube API)
    Params:
    
    youtube: the build object from googleapiclient.discovery
    video_ids: list of video IDs
    
    Returns:
    Dataframe with video IDs and associated top level comment in text.
    
    """
    all_comments = []
    
    for video_id in video_ids:
        try:   
            request = youtube.commentThreads().list(
                part="snippet,replies",
                videoId=video_id
            )
            response = request.execute()
        
            comments_in_video = [comment['snippet']['topLevelComment']['snippet']['textOriginal'] for comment in response['items'][0:10]]
            comments_in_video_info = {'video_id': video_id, 'comments': comments_in_video}

            all_comments.append(comments_in_video_info)
            
        except: 
            # When error occurs - most likely because comments are disabled on a video
            print('Could not get comments for video ' + video_id)
        
    return pd.DataFrame(all_comments)     
```

## Preprocessing and Feature Engineering
- Create publish day (in the week) column
- Convert needed columns from object to int
- Drop unwanted columns
- Convert duration to seconds
- Extract Length of video title
- Extract Experience_years (how many years channel created)
- Extract total video likes for each channel
- first_upload_year (How many years has it been since the first video was uploaded?)

```python
# Create publish day (in the week) column
videos_df['publishedAt'] =  videos_df['publishedAt'].apply(lambda x: parser.parse(x)) 
videos_df['pushblishDayName'] = videos_df['publishedAt'].apply(lambda x: x.strftime("%A")) 
```

```python
# Convert needed columns from object to int 
numeric_columns = ['viewCount', 'likeCount', 'commentCount']
videos_df[numeric_columns] = videos_df[numeric_columns].apply(pd.to_numeric, errors='coerce', axis=1)
```

```python
# Drop unwanted columns
videos_df = videos_df.drop(['caption','favouriteCount'],axis=1)
```

```python
# Convert duration to seconds
videos_df['durationSecs'] = videos_df['duration'].apply(lambda x: isodate.parse_duration(x))
videos_df['durationSecs'] = videos_df['durationSecs'].astype('timedelta64[s]')
videos_df['durationSecs'] = videos_df['durationSecs'].astype('int64')
videos_df = videos_df.drop(['duration'],axis=1)
```

```python
# Extract Length of video title  
videos_df['titleLength'] = videos_df['title'].apply(lambda x: len(x))
```

```python
# experience_years (how many years channel published)
channels_df['publishedDate'] =  channels_df['publishedDate'].apply(lambda x: parser.parse(x)) 
channels_df['experience_years'] = 2022 - channels_df['publishedDate'].dt.year
```

```python
# Total video likes of each channel 
channels_df['totalLikes'] = videos_df['likeCount'].groupby(videos_df['channelTitle']).transform('sum').unique()
channels_df['totalLikes'] =  channels_df['totalLikes'].apply(lambda x:int(x)) 
```

```python
# first_upload_year (How many years has it been since the first video was uploaded?)

first_upload_years=videos_df.loc[videos_df.groupby('channelTitle')['publishedAt'].idxmin()][['channelTitle','publishedAt']]
first_upload_years['first_upload_years']=2022-first_upload_years.loc[videos_df.groupby('channelTitle')['publishedAt'].idxmin()]['publishedAt'].dt.year
first_upload_years = first_upload_years.rename({'channelTitle':'channelName'}, axis=1)
first_upload_years = first_upload_years.drop(['publishedAt'],axis=1)
channels_df = pd.merge(channels_df, first_upload_years, how='left', on='channelName')
```

```python
# this urls from text
def remove_url(text):
    url = re.compile(r'https?://\S+|www\.\S+')
    return url.sub(r'',str(text))

# this function remove punct like (<]./xxs)
def remove_punct(text):
    table = str.maketrans('', '', string.punctuation)
    return text.translate(table)

 # this function remove html tags   
def remove_html(text):
    if text != "":
        html=re.compile(r'<.*?>')
        return html.sub(r'',str(text))
```

```python
# clean comments for wordcloud
stop_words = set(stopwords.words('english'))
comments_df['clear_comments'] = comments_df['comments'].apply(lambda x: [item for item in str(x).split() if item not in stop_words])
comments_df['clear_comments'] = comments_df['clear_comments'].apply(lambda x: remove_url(x))
comments_df['clear_comments'] = comments_df['clear_comments'].apply(lambda x: remove_punct(x))
comments_df['clear_comments'] = comments_df['clear_comments'].apply(lambda x: remove_html(x))
comments_df =  comments_df.dropna()
```

```python
# clean titles for wordcloud
stopwords = set(stopwords.words('english'))
videos_df['clear_titles'] = videos_df['title'].apply(lambda x: [item for item in str(x).split() if item not in stopwords])
videos_df['clear_titles'] = videos_df['clear_titles'].apply(lambda x: remove_url(x))
videos_df['clear_titles'] = videos_df['clear_titles'].apply(lambda x: remove_punct(x))
videos_df['clear_titles'] = videos_df['clear_titles'].apply(lambda x: remove_html(x))
```

## Sentiment analysis for the comments of each channel 
This stage was a lot of fun for me and had a lot of potential. I could analyze the sentiments of the most viewed videos or the least viewed videos, but I decided to analyze the comments of all the videos of each channel.

How : first i separated the videos of the channel and put the comments of each video cleanly into a presentation using text processing and used the VADER model from nltk to analyze sentiments.


```python
def remove_stopwords(data):  
    """
    Remove stopwords from each string element of list
    Params:
    
    list_of_strings: list of comments
    
    Returns:
    cleaned list without stopwords
    """
    output_array=[]
    for sentence in data:
        temp_list=[]
        for word in sentence.split():
            if word.lower() not in stopwords:
                temp_list.append(word)
        output_array.append(' '.join(temp_list))
    return output_array

def percentage(part, whole):
    """
    This funciton will calculate percentage i used it to see 
    percentage of each sentiment for each channel

    Params:
    part: part number
    whole: whole number 

    Returns:
    percentage %
    """
    Percentage = 100 * float(part)/float(whole)
    return str(Percentage) + "%"

def sentiment_analysis(list_of_comments):
    """
    Apply sentimentInt analysis for coemments of each channel  
    Params:
    
    list_of_comments: list of comments
    
    Returns:
    Dictionary which contains the number of (positive, natural and negative) comments of each channel 
    """
    analyzer = SentimentIntensityAnalyzer()
    listof = []
    
    # split comments by (,) for each video 
    list_of_all_comments = [item.strip() for sublist in [element.split(",") for element in list_of_comments] for item in sublist]
    
    # remove url from comments
    list_of_all_comments2 = [remove_url(item) for item in list_of_all_comments]
    
    # remove html from comments
    list_of_all_comments3 = [remove_html(item) for item in list_of_all_comments2]
    
    # remove punc from comments
    list_of_all_comments4 = [re.sub(r'[^\w\s]', ' ', str(item)) for item in list_of_all_comments3]
    
    # delete extra space 
    list_of_all_comments5 = [" ".join(item.split()) for item in list_of_all_comments4]
    
    # lower case all letters in comments 
    list_of_all_comments6 = [item.lower() for item in list_of_all_comments5]
    
    # remove stopwords in comments 
    list_of_all_comments7 = remove_stopwords(list_of_all_comments6)
    
    for i in list_of_all_comments7 : 
        sentemnts = analyzer.polarity_scores(i)
        if sentemnts['compound'] >= 0.05 :
            a = "Positive"
            listof.append(a)
        elif sentemnts['compound'] <= - 0.05 :
            a = "Negative"
            listof.append(a)
        else :
            a = "Neutral"
            listof.append(a)

    res = {i:listof.count(i) for i in listof}
    return res
```

```python
# extract comments of each video of each channel
codebasics_comments = comments_df[comments_df['channelName']=='codebasics']['comments'].to_list()
sentdex_comments = comments_df[comments_df['channelName']=='sentdex']['comments'].to_list()
KrishNaik_comments = comments_df[comments_df['channelName']=='Krish Naik']['comments'].to_list()
KenJee_comments = comments_df[comments_df['channelName']=='Ken Jee']['comments'].to_list()
```

```python
# apply sentiment_analysis
result_codebasics = sentiment_analysis(codebasics_comments)
result_KrishNaik = sentiment_analysis(KrishNaik_comments)
result_KenJee = sentiment_analysis(KenJee_comments)
result_sentdex = sentiment_analysis(sentdex_comments)
```
